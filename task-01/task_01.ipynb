```python
import pandas as pd


from google.colab import files
uploaded = files.upload()


df = pd.read_csv("data.csv")

# Displaying the first 5 rows
print("--- Head (First 5 rows) ---")
print(df.head())

# Display the last 5 rows
print("\n--- Tail (Last 5 rows) ---")
print(df.tail())

# Displaying data types
print("\n--- Data Types ---")
print(df.dtypes)

# Displaying summary statistics
print("\n--- Summary Statistics (Data_value) ---")
print(df["Data_value"].describe())

# Calculating individual statistics
mean_val = df["Data_value"].mean()
median_val = df["Data_value"].median()
min_val = df["Data_value"].min()
max_val = df["Data_value"].max()
count_val = df["Data_value"].count()

print("\nSpecific Stats for 'Data_value':")
print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Min: {min_val}")
print(f"Max: {max_val}")
print(f"Count: {count_val}")

# Filtering rows
filtered_df = df[df["Series_title_1"] == "Sales (operating income)"]


selected_columns = filtered_df[
    ["Period", "Data_value", "Series_title_1", "STATUS"]
]

# Slicing first 50 rows
subset_slice = selected_columns.iloc[:50]

print("\n--- Filtered & Sliced Data (First 5 rows) ---")
print(subset_slice.head())

# Saving filtered data to a new CSV file
output_filename = "filtered_sales_data.csv"
subset_slice.to_csv(output_filename, index=False)

print(f"\nProject complete: Filtered results saved to '{output_filename}'")
```
